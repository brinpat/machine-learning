```python
data = 'https://www.doogal.co.uk/AdministrativeAreasCSV/?district=E08000029'
postcodes = pd.read_csv(data)
```

```python


postcodes['Postcode'].str.split(' ', n=1, expand=True)[0] + '%20' + postcodes['Postcode'].str.split(' ', n=1, expand=True)[1]
```

```python
postcodes['Metadata'] = 'https://www.doogal.co.uk/ShowMap?postcode=' + postcodes['Postcode'].str.split(' ', n=1, expand=True)[0] + '%20' + postcodes['Postcode'].str.split(' ', n=1, expand=True)[1] 
```

```python
postcodes[postcodes['Postcode'] == 'B26 3AW']
```

```python
color_scale = [('Yes', 'orange'), ('No','red')]

fig = px.scatter_mapbox(postcodes, 
                        lat="Latitude", 
                        lon="Longitude", 
                        hover_name="Postcode", 
                        text = 'Metadata',
                        hover_data=['Postcode', 'Ward', "Parish", 'Introduced', 'Altitude'],
                        color="In Use?",
                        color_continuous_scale=color_scale,
                        zoom=8, 
                        height=800,
                        width=800)

fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})


fig.show()
```

```python
import folium
import webbrowser
```

```python
map1 = folium.Map(location=[np.mean(postcodes['Latitude']),np.mean(postcodes['Longitude'])], zoom_start=14)
```

```python
i = 0

for _, location in postcodes.iterrows():
    
    tooltip = [postcodes['Postcode'][i], postcodes['Ward'][i], postcodes['Parish'][i], postcodes['Introduced'][i], postcodes['Altitude'][i]]
    
    locations = [location['Latitude'],location['Longitude']]
    
    pop = '<a href=' + postcodes['Metadata'][i] + ' target="_blank">' + postcodes['Postcode'][i] + '</a>'

    popup = pop

    marker = folium.Marker(location = locations,popup = popup, tooltip = tooltip, icon=folium.Icon(color="blue", icon="home")).add_to(map1)
    
    i+=1
    
    
map1.save("solihull_postcodes.html")
```
